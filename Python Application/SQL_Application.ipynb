```python
def execute_query(connection, query):
	cursor = connection.cursor()
	try:
		cursor.execute(query)
		print("Query Executed Successfully\n")
		cols = [i[0] for i in cursor.description]
		res = cursor.fetchall()
		return res, cols
	except Error as err:
		print(err)
        
```

```python
import pandas as pd
df_course = pd.read_sql_query("select * from course_table", con=connection)
df_acadamics_table = pd.read_sql_query("select * from acadamics_table", con=connection)
df_admission_table = pd.read_sql_query("select * from admission_table", con=connection)
df_department_table = pd.read_sql_query("select * from department_table", con=connection)
df_enrollment_table = pd.read_sql_query("select * from enrollment_table", con=connection)
df_extra_curricular_activities_table = pd.read_sql_query("select * from extra_curricular_activities_table", con=connection)
df_faculty_table = pd.read_sql_query("select * from faculty_table", con=connection)
df_project_table = pd.read_sql_query("select * from project_table", con=connection)
df_student_table = pd.read_sql_query("select * from student_table", con=connection)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# This plot will give the students and their respective GPA

sns.stripplot(x ='Student_ID', y ='GPA', data = df_acadamics_table)
```

```python
# This plot explains about the types of project funding accross various domains

sns.stripplot(x ='Project_Funding', y ='Project_Domain', data = df_project_table,
              jitter = True, dodge = True)
```

```python
# This plot tells about the type of student admissions and number of total admissions falling under the same

ax = df_admission_table['Admission_Type'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Type of Admission vs students table")
ax.set_xlabel("Admission Type")
ax.set_ylabel("Number of Students")
```
